**Support Vector Machine (SVM)**

SVM is a supervised machine learning algorithms used for classification and regression tasks. The main objective is to find an optimal hyperplane that best separates data into different classes in high-dimensional space. The hyperplane is chosen to maximize the margin, which is the distance between the hyperplane and the nearest data points (support vectors) of each class. Kernel functions are used for transforming lower dimensional input space to higher dimensional output. SVM is effective in high-dimensional spaces, with large number of features and non-linear classification. It also can provide probability estimates for classification tasks. However, SVM can be computationally expensive and may not be suitable for large datasets. Also, since the data points are placed above and below the hyperplane there is no probabilistic explanation for this algorithm.  For predicting customer purchase behavior, SVM can classify using the features with a radial basis function (RBF) kernel to capture non-linear relationships.

**K-Nearest Neighbors (KNN)**

KNN is a simple and very effective supervised machine learning algorithm. It makes predictions based on the similarity of input data points. The performance of KNN is basically based on the choice of the value of K. KNN works by memorizing the entire training dataset. When a new data point is given for prediction, KNN looks at the k-nearest data points in the training set based on a specified distance metric (commonly Euclidean distance).
KNN is simple and versatile with no assumption about the data distribution (non-parametric method). However, the performance of KNN can be degraded from outliers and noisy data. It is computationally expensive as it requires storing the entire dataset and calculating distances for each query. There is a performance drop when it comes to high-dimensional data and large datasets.
KNN can predict customer purchase behavior, by identifying similar past customers based on proximity in feature space.

**Hybrid Model using Ensemble Learning**

In Ensemble Learning, results from multiple models are combined to get a more precise prediction and improve overall performance. It can be observed that the accuracy and resilience in prediction can be improved massively by using Ensemble machine learning techniques. The errors and biases that exist in individual models are mitigated by this approach. It is flexible as it can use a combination of many algorithms such as SVM, KNN etc. However, it requires more computation and careful model selection. It is hard to interpret the results due to model blending. The author proposes implementing a hybrid model using ensemble learning combining SVM and KNN as they will give the best outperform other techniques.

| Technique | Strengths | Advantages | Weaknesses | Disadvantages | Input Data | Output Data |
| --- | --- | --- | --- | --- | --- | --- |
| SVM | Effective for high-dimensional data | Effective for non-linear relationships | No probablistic explanation  | Computationally expensive | Scaled and Preprocessed Numerical/Categorical data | Binary class labels |
| KNN | Simple, Versatile and non-parametric | Adapts to new data easily | Sensitive to outliers and noisy data  | Computationally eexpensive for large datasets | Scaled and Preprocessed Numerical/Categorical data | Binary class labels |
| Hybrid Model | Robust and flexible | Improves accuracy and resilience | Complex to implement and tune  | Difficult to interpret due to model combination | Scaled and Preprocessed Numerical/Categorical data | Binary Class label based on the technique selected |

```python
# import libraries
from sklearn.ensemble import VotingClassifier
```

```python
# Combining SVM and KNN as they gave the best results

# Training the SVM model
# Probability set to True for soft-voting
svm_model = SVC(kernel='rbf',C=1, random_state=1, probability=True)
# Training the KNN Model
knn_model = KNeighborsClassifier(n_neighbors=5)

# Training the Hybrid Model using Voting Classifier (Ensemble Learning Technique) from sklearn library
vc_model = VotingClassifier(estimators=[('svm', svm_model), ('knn', knn_model)], voting='soft')
vc_model.fit(X_train, Y_train)

# Prediciting using VC Hybrid Model (Binary class labels prediction)
vc_pred = vc_model.predict(X_test)
```

```python
# Prediciting using VC Hybrid Model (Probability based prediction)
vc_proba = vc_model.predict_proba(X_test)
```

#### **ROC AUC Scores and Curves**

ROC AUC score is a popular metric to evaluate the performance of binary classifiers. The ROC curve shows the performance of a binary classifier with different decision thresholds. It plots the True Positive Rates (TPR) against the False Positive Rates (FPR). The ROC AUC score is the area under the ROC curve. It sums up how well a model can produce relative scores to discriminate between positive or negative instances across all classification thresholds. The ROC AUC score ranges from 0 to 1, where 0.5 indicates random guessing, and 1 indicates perfect performance.A Higher ROC AUC score indicates better performance.

| Technique                         | ROC AUC Score (Binary Class Prediction) | ROC AUC Score (Binary Class Prediction) |
| ---------------------------------- | -------- | -------- |
| SVM                                | 0.895  | 0.969 |
| KNN                                | 0.913  | 0.960 |
| Hybrid Model (Voting Classifier)   | 0.923   | 0.977 |
| Hybrid Model (Stacking)            | 0.932   | 0.977 |

```python
def plot_sklearn_roc_curve(model_data, Y_test):
  num_models = len(model_data)
  fig = plt.figure(figsize = (15, 5*num_models))
  for i, (model_name, model_pred, model_proba) in enumerate(model_data):
    # ROC curve using predictions of binary class labels - bad plot
    fpr1, tpr1, _ = roc_curve(Y_test, model_pred)
    # ROC curve using predictions based on probabilities for each class - good plot
    fpr2, tpr2, _ = roc_curve(Y_test, model_proba)
    # Creating subplots for both ROC curves
    ax1 = plt.subplot(num_models,2,2*i+1)
    RocCurveDisplay(fpr=fpr1, tpr=tpr1).plot(ax = ax1)
    ax2 = plt.subplot(num_models,2,2*i+2)
    RocCurveDisplay(fpr=fpr2, tpr=tpr2).plot(ax = ax2)
  fig.subplots_adjust(hspace=0.3, wspace=0.2)
  plt.show()
```

```python
# tuple with model data such as the name, predictions of class labels, predictions based on probabilities for each class for each model
model_data = [
    ('SVM', svm_pred, svm_proba[:,1]),
    ('KNN', knn_pred, knn_proba),
    ('VC', vc_pred, vc_proba[:,1]),
    ('Stacking', stacking_pred, stacking_proba[:,1])
]
```

```python
# Visualising ROC curves for each model
plot_sklearn_roc_curve(model_data, Y_test)
```

```python
# All ROC Curves from model_proba in one plot
def plot_all_roc_curve(model_data, Y_test):
  plt.figure(figsize = (10, 5))
  for i, (model_name,model_pred, model_proba) in enumerate(model_data):
    fpr, tpr, _ = roc_curve(Y_test, model_proba)
    plt.plot(fpr, tpr, label=model_name)
  plt.title("All ROC Curves")
  plt.xlabel("False Positive Rate")
  plt.ylabel("True Positive Rate")
  plt.legend(loc='lower right')
  plt.show()

# Visualizing all ROC curves of each model in one plot
plot_all_roc_curve(model_data, Y_test)
```

It can be observed through the comparison of ROC AUC scores, the Hybrid models have values significantly more closer to 1, indicating models are good at seperating classes and making better predictions. The ROC curves also indicate it is a good model without much overlap.
